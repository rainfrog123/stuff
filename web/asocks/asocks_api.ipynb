```python
import requests
import json
import time
import hashlib
import hmac
import base64

# API Key
API_KEY = 'eV3wFrpclIBrZYuaCwxXBZ1oJDM2r0Kw'
BASE_URL = 'http://api.asocks.com/v2'

# Generate signature (if required for additional security features)
def generate_signature(secret_key, http_method, url, timestamp, body=''):
    data = f"{timestamp}{http_method.upper()}{url}{body}"
    signature = hmac.new(secret_key.encode('utf-8'), data.encode('utf-8'), hashlib.sha256).digest()
    return base64.b64encode(signature).decode()

# Example API call functions

# Get user balance
def get_user_balance():
    url = f"{BASE_URL}/user/balance?apiKey={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# Get tariff information
def get_tariff_info():
    url = f"{BASE_URL}/plan/info?apiKey={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# Create a new proxy port
def create_proxy_port():
    url = f"{BASE_URL}/proxy/create-port?apiKey={API_KEY}"
    response = requests.post(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# List all proxy ports
def list_proxy_ports():
    url = f"{BASE_URL}/proxy/ports?apiKey={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# Refresh a proxy port's IP
def refresh_proxy_ip(port_id):
    url = f"{BASE_URL}/proxy/refresh/{port_id}?apiKey={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# Example usage
user_balance = get_user_balance()
tariff_info = get_tariff_info()
proxy_ports = list_proxy_ports()

# Display the results
if user_balance:
    print("User Balance:", user_balance)

if tariff_info:
    print("Tariff Information:", tariff_info)

if proxy_ports:
    print("Proxy Ports:", proxy_ports)

# Function to save the notebook
# import ace_tools as tools; tools.display_ipynb_to_user('Asocks_API_Example')
```

```python
import requests

def get_bin_data(bin_id):
    try:
        response = requests.get(f'https://data.handyapi.com/bin/{bin_id}', headers={'Referer': 'your-domain'})
        if response.status_code == 200:
            data = response.json()
            return {
                'Type': data.get('Type'),
                'Issuer': data.get('Issuer'),
                'CardTier': data.get('CardTier'),
                'CountryName': data.get('Country', {}).get('Name')
            }
        else:
            return {'Type': None, 'Issuer': None, 'CardTier': None, 'CountryName': None}
    except Exception as e:
        return {'Type': None, 'Issuer': None, 'CardTier': None, 'CountryName': None}


get_bin_data(535316)
```
