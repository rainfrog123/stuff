```python
import requests
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import sys

def start_adspower_browser(ads_id):
    open_url = f"http://local.adspower.com:50325/api/v1/browser/start?user_id={ads_id}"
    close_url = f"http://local.adspower.com:50325/api/v1/browser/stop?user_id={ads_id}"

    response = requests.get(open_url).json()
    print("AdsPower API response:", response)  # Debug statement
    if response["code"] != 0:
        print(response["msg"])
        print("Please check ads_id")
        sys.exit()

    chrome_driver_path = response["data"]["webdriver"]
    selenium_ws_url = response["data"]["ws"]["selenium"]

    return chrome_driver_path, selenium_ws_url, close_url

def initialize_webdriver(chrome_driver_path, selenium_ws_url):
    service = Service(executable_path=chrome_driver_path)
    chrome_options = Options()
    chrome_options.add_experimental_option("debuggerAddress", selenium_ws_url)
    
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

# Initialize the browser and WebDriver
ads_id = "kkomyls"
chrome_driver_path, selenium_ws_url, close_url = start_adspower_browser(ads_id)

print("Chrome Driver Path:", chrome_driver_path)  # Debug statement
print("WebSocket URL:", selenium_ws_url)  # Debug statement

driver = initialize_webdriver(chrome_driver_path, selenium_ws_url)

# Now the browser window is open and you can use it as your main window
try:
    driver.get("https://www.google.com")
    print(driver.title)
except Exception as e:
    print(f"An error occurred: {e}")

# You can keep interacting with the driver in your notebook cells
```

```python
# query
import requests

# Define the API endpoint
api_url = "http://localhost:50325/api/v1/group/list"

# Define the parameters for the request
params = {
    "group_name": "",  # Leave empty to search all groups
    "page": 1,         # Default is page 1
    "page_size": 10    # Default is 10, max is 2000
}

# Send the GET request
response = requests.get(api_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    if data['code'] == 0:
        print("Query successful.")
        print("Groups data:")
        for group in data['data']['list']:
            print(f"Group ID: {group['group_id']}, Group Name: {group['group_name']}, Remark: {group['remark']}")
    else:
        print(f"Query failed: {data['msg']}")
else:
    print(f"Request failed with status code: {response.status_code}")

```

```python
import requests

# Define the API endpoint for creating a group
create_api_url = "http://localhost:50325/api/v1/group/create"

# Define the parameters for the create request
create_params = {
    "group_name": "new_group",  # Name of the new group
    "remark": "This is a new group"  # Remark for the new group
}

# Send the POST request to create a new group
response = requests.post(create_api_url, json=create_params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    if data['code'] == 0:
        print("Group creation successful.")
        print(f"New Group ID: {data['data']['group_id']}")
    else:
        print(f"Group creation failed: {data['msg']}")
else:
    print(f"Request failed with status code: {response.status_code}")
```
