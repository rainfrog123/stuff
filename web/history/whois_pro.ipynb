```python
import socket
import itertools
import string

def is_domain_available(domain):
    """
    Check if a given domain is available for registration using DNS queries.

    Args:
        domain (str): The domain name to check.

    Returns:
        bool: True if the domain is available, False otherwise.
    """
    try:
        # Try to resolve the domain name
        socket.gethostbyname(domain)
        return False  # If resolution is successful, the domain is taken
    except socket.gaierror:
        return True  # If resolution fails, the domain might be available

def generate_domains():
    """
    Generate domain names of the format xxyy.pro, where x and y are lowercase letters.

    Yields:
        str: A domain name string.
    """
    letters = string.ascii_lowercase  # a-z
    for first_pair, second_pair in itertools.product(letters, repeat=2):
        yield f"{first_pair}{first_pair}{second_pair}{second_pair}.pro"

def main():
    """
    Main function to check the availability of generated domains
    and save the available ones to a file.
    """
    available_domains = []

    # Open the output file to write available domains
    with open('available_domains.txt', 'w') as file:
        # Iterate over each generated domain
        for domain in generate_domains():
            if is_domain_available(domain):
                available_domains.append(domain)
                file.write(f"{domain}\n")
                print(f"Available: {domain}")
            else:
                print(f"Taken: {domain}")

    print(f"\nTotal available domains found: {len(available_domains)}")

if __name__ == "__main__":
    main()
```
