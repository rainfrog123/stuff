```python
import whois
import itertools
import string

def check_domain_availability(domain):
    try:
        w = whois.whois(domain)
        if w.domain_name is None:
            return True
    except Exception as e:
        if any(keyword in str(e) for keyword in ["No match for domain", "NOT FOUND"]):
            return True
    return False

def generate_domains():
    letters = string.ascii_lowercase  # a-z
    # Generate domain combinations of the form xxyy.pro
    for first_pair, second_pair in itertools.product(letters, repeat=2):
        domain = f"{first_pair}{first_pair}{second_pair}{second_pair}.pro"
        yield domain

def main():
    available_domains = []
    with open('available_domains.txt', 'w') as file:
        for domain in generate_domains():
            if check_domain_availability(domain):
                available_domains.append(domain)
                file.write(domain + '\n')
                print(f"Available: {domain}")
            else:
                print(f"Taken: {domain}")

    print(f"\nTotal available domains found: {len(available_domains)}")

if __name__ == "__main__":
    main()
```
