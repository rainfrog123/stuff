```python
import requests
import json

# Define your API key and the base URL for the API
API_KEY = 'e102bd9aedc76786a931f6afa1b1efe4'  # Replace with your actual API key
API_URL = 'http://localhost:50325/api/v1/user/list'

# Set the headers for the API request
headers = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {API_KEY}'
}

# Make a GET request to the API to retrieve the list of profiles
response = requests.get(API_URL, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    profiles = response.json()
    # Pretty print the entire response JSON
    print("Response JSON:")
    print(json.dumps(profiles, indent=4))  # Beautify the JSON response
    
    print("\nProfile IDs:")
    # Adjust based on actual structure
    if 'data' in profiles and 'list' in profiles['data']:
        for profile in profiles['data']['list']:
            print(f"Profile Name: {profile.get('name', 'N/A')}, Profile ID: {profile.get('user_id', 'N/A')}")
    else:
        print("Key 'data' or 'list' not found in the response.")
else:
    print("Failed to retrieve profiles")
    print(json.dumps(response.json(), indent=4))  # Beautify the error response
```

```python
import requests
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import sys

def start_adspower_browser(ads_id):
    open_url = f"http://local.adspower.com:50325/api/v1/browser/start?user_id={ads_id}"
    close_url = f"http://local.adspower.com:50325/api/v1/browser/stop?user_id={ads_id}"

    response = requests.get(open_url).json()
    print("AdsPower API response:", response)  # Debug statement
    if response["code"] != 0:
        print(response["msg"])
        print("Please check ads_id")
        sys.exit()

    chrome_driver_path = response["data"]["webdriver"]
    selenium_ws_url = response["data"]["ws"]["selenium"]

    return chrome_driver_path, selenium_ws_url, close_url

def initialize_webdriver(chrome_driver_path, selenium_ws_url):
    service = Service(executable_path=chrome_driver_path)
    chrome_options = Options()
    chrome_options.add_experimental_option("debuggerAddress", selenium_ws_url)
    
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

# Initialize the browser and WebDriver
ads_id = "kkomyls"
chrome_driver_path, selenium_ws_url, close_url = start_adspower_browser(ads_id)

print("Chrome Driver Path:", chrome_driver_path)  # Debug statement
print("WebSocket URL:", selenium_ws_url)  # Debug statement

driver = initialize_webdriver(chrome_driver_path, selenium_ws_url)

# Now the browser window is open and you can use it as your main window
try:
    driver.get("https://www.google.com")
    print(driver.title)
except Exception as e:
    print(f"An error occurred: {e}")

# You can keep interacting with the driver in your notebook cells
```
