```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the neural network architecture
ax.scatter([0.5] * 3, np.arange(1, 4), color='blue', s=500, label='Input Layer')
ax.scatter([1.5, 1.5], [1, 2], color='green', s=500, label='Hidden Layer')
ax.scatter([2.5], [1], color='red', s=500, label='Output Layer')

# Connect the neurons with lines
ax.plot([0.5, 1.5], [1, 1], color='gray')
ax.plot([0.5, 1.5], [2, 1], color='gray')
ax.plot([1.5, 2.5], [1, 1], color='gray')

# Add labels and legend
ax.set_xticks([0.5, 1.5, 2.5])
ax.set_xticklabels(['Input', 'Hidden', 'Output'])
ax.set_yticks([])
ax.legend()

# Add text annotations for forward pass, backward pass, and gradients
ax.annotate('Forward Pass', xy=(0.5, 1.5), xytext=(0.7, 1.5), arrowprops=dict(arrowstyle='->'))
ax.annotate('Backward Pass', xy=(2.5, 1.5), xytext=(2.3, 1.5), arrowprops=dict(arrowstyle='->'))

# Text annotations for gradient computation
ax.text(1.5, 2.2, r'$\delta^{(2)}_1$', fontsize=14, color='black')
ax.text(1.5, 0.8, r'$\delta^{(2)}_2$', fontsize=14, color='black')

ax.text(1.2, 3, r'$\frac{\partial C}{\partial w^{(2)}_1}$', fontsize=14, color='black')
ax.text(1.8, 3, r'$\frac{\partial C}{\partial w^{(2)}_2}$', fontsize=14, color='black')

ax.text(1.2, 1.3, r'$\frac{\partial C}{\partial b^{(2)}}$', fontsize=14, color='black')

ax.text(0.2, 1.5, r'$a^{(1)}_1$', fontsize=14, color='black')
ax.text(0.2, 0.5, r'$a^{(1)}_2$', fontsize=14, color='black')
ax.text(1.7, 1.5, r'$a^{(2)}_1$', fontsize=14, color='black')
ax.text(1.7, 0.5, r'$a^{(2)}_2$', fontsize=14, color='black')
ax.text(2.7, 1.5, r'$a^{(3)}_1$', fontsize=14, color='black')

# Show the plot
plt.show()
```
