```python
from binance.client import Client
import time

# Replace with your Binance API key and secret
API_KEY = key
API_SECRET = secret

# Initialize Binance client
client = Client(API_KEY, API_SECRET)

def set_max_leverage():
    try:
        # Fetch all futures trading pairs
        futures_exchange_info = client.futures_exchange_info()
        symbols = [symbol['symbol'] for symbol in futures_exchange_info['symbols'] if symbol['contractType'] == 'PERPETUAL']

        # Loop through each pair and set the maximum leverage
        for symbol in symbols:
            try:
                # Fetch leverage brackets for the pair
                leverage_info = client.futures_leverage_bracket()
                for item in leverage_info:
                    if item['symbol'] == symbol:
                        max_leverage = max([bracket['initialLeverage'] for bracket in item['brackets']])

                        # Set the leverage to max for this pair
                        response = client.futures_change_leverage(symbol=symbol, leverage=max_leverage)
                        print(f"Set max leverage ({max_leverage}) for {symbol}: {response}")
                        break
                
                # Pause to avoid hitting API limits
                time.sleep(0.2)  # Adjust delay as necessary
            except Exception as e:
                print(f"Failed to set leverage for {symbol}: {e}")

    except Exception as e:
        print(f"Error fetching exchange info: {e}")

# Execute the function
set_max_leverage()
```

```python
from binance.client import Client

# Replace with your Binance API key and secret
API_KEY = "your_api_key"
API_SECRET = "your_api_secret"

# Initialize Binance client
client = Client(API_KEY, API_SECRET)

def get_non_tradeable_perp_pairs_with_status():
    try:
        # Fetch all futures exchange information
        futures_exchange_info = client.futures_exchange_info()
        
        # Get symbols and their respective statuses if not trading
        non_trading_pairs = [
            {"symbol": symbol['symbol'], "status": symbol['status']}
            for symbol in futures_exchange_info['symbols']
            if symbol['contractType'] == 'PERPETUAL' and symbol['status'] != 'TRADING'
        ]
        
        return non_trading_pairs

    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Get non-trading pairs with their status
non_trading_pairs = get_non_tradeable_perp_pairs_with_status()

print("\nNon-Trading Pairs with Status:")
for pair in non_trading_pairs:
    print(f"Symbol: {pair['symbol']}, Status: {pair['status']}")
```
